```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dataset=pd.read_csv("/content/Market_Basket_Optimisation.csv")
```

```python
dataset.head(10)
```

```python
dataset.shape
```

```python
items=[]
for i in range(0,7499):
  items.append([str(dataset.values[i,j])
    for j in range(0,20)])
```

```python
items[0]
```

```python
Assn_rules=apriori(transactions=items,
                   min_support=0.003,
                   min_confidence=0.04,
                   min_lift=3,
                   min_length=2,
                   max_length=3)
```

```python
Results=list(Assn_rules)
```

```python
Results
```

```python
def inspect(Results):
    lhs=[tuple(result[2][0][0])[0] for result in Results]
    rhs=[tuple(result[2][0][1])[0] for result in Results]
    support=[result[1] for result in Results]
    confidences=[result[2][0][2] for result in Results]
    lifts= [result[2][0][3] for result in Results]
    return list(zip(lhs,rhs,support,confidences,lifts))
dataframe_final=pd.DataFrame(inspect(Results),columns=['Left Hand Side','Right Hand Side','Support','Confidence',
                                                    'Lift'])
```
